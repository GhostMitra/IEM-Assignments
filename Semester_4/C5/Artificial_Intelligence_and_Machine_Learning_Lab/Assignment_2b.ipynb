### Question 1
Create the Employee DataFrame The employee dataset contains the following columns:

Emp_ID, Name, Age, Department, Designation, Salary, Experience, City

The dataset consists of 10 employee records.

```python
import pandas as pd

# ---------- SAMPLE DATASET (ONLY FOR CREATION) ----------
data = {
    "Emp_ID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Amit","Riya","Rahul","Sneha","Karan","Neha","Ankit","Pooja","Rohit","Simran"],
    "Age": [28,26,30,27,29,25,31,28,26,27],
    "Department": ["IT","HR","IT","Finance","IT","HR","Finance","IT","HR","IT"],
    "Designation": ["Engineer","HR Exec","Developer","Analyst","Manager",
                    "HR Exec","Accountant","Developer","HR Exec","Engineer"],
    "Salary": [60000,45000,70000,50000,80000,42000,48000,65000,46000,72000],
    "Experience": [5,3,6,4,8,2,5,6,3,7],
    "City": ["Kolkata","Delhi","Mumbai","Pune","Bangalore",
             "Delhi","Chennai","Hyderabad","Pune","Bangalore"]
}


df = pd.DataFrame(data)
print("\nEMPLOYEE DATAFRAME:\n", df)
```

### Question 2
Display first 5 records

```python
# ---------- FIRST N RECORDS ----------
n = int(input("\nEnter number of records to display from start: "))
print(df.head(n))
```

### Question 3
Display the last 3 records.

```python
# ---------- LAST N RECORDS ----------
m = int(input("\nEnter number of records to display from end: "))
print(df.tail(m))
```

### Question 4
Find shape of DataFrame

```python
# ---------- SHAPE ----------
print("\nShape of DataFrame:", df.shape)
```

### Question 5
Display column names

```python
# ---------- COLUMN NAMES ----------
print("\nColumn Names:")
print(list(df.columns))
```

### Question 6
Check missing values

```python
# ---------- MISSING VALUES ----------
print("\nMissing Values in each column:")
print(df.isnull().sum())
```

### Question 7
Find average salary

```python
# ---------- AVERAGE SALARY ----------
print("\nAvailable Columns:", list(df.columns))
sal_col = input("Enter salary column name: ")
print("Average Salary:", df[sal_col].mean())
```

### Question 8
Print the employee information where Salary greater than 50000

```python
# ---------- SALARY CONDITION ----------
value = int(input("\nEnter salary value to filter greater than: "))
print(df[df[sal_col] > value])
```

### Question 9
Count IT Department employees

```python
# ---------- COUNT IT DEPARTMENT ----------
dept_name = input("\nEnter department name to count: ")
print("Employee Count:", (df["Department"] == dept_name).sum())
```

### Question 10
Print Employee count per department

```python
# ---------- EMPLOYEE COUNT PER DEPARTMENT ----------
print("\nEmployee count per department:")
print(df["Department"].value_counts())
```

### Question 11
Print Max and Min salary

```python
# ---------- MAX & MIN SALARY ----------
print("\nMaximum Salary:", df[sal_col].max())
print("Minimum Salary:", df[sal_col].min())
```

### Question 12
Save DataFrame to CSV

```python
# ---------- SAVE TO CSV ----------
file_name = input("\nEnter CSV file name to save (with .csv): ")
df.to_csv(file_name, index=False)
print("Data saved successfully")
```
