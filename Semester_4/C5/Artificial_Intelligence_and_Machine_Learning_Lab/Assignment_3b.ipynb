### Question 1
Create an employee dataset with missing values.

```python
import pandas as pd
import numpy as np

data = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Aman', 'Riya', 'Kunal', 'Neha', 'Arjun'],
    'Age': [25, np.nan, 30, 28, np.nan],
    'Department': ['HR', 'IT', np.nan, 'Finance', 'IT'],
    'Salary': [50000, 60000, np.nan, 55000, np.nan],
    'Experience': [2, 4, np.nan, 3, 5]
}

df = pd.DataFrame(data)
```

### Question 2
Display the dataset.

```python
df
```

### Question 3
Identify missing values using isnull().

```python
df.isnull()
```

### Question 4
Count missing values column-wise.

```python
df.isnull().sum()
```

### Question 5
Find the percentage of missing values in each column.

```python
(df.isnull().sum() / len(df)) * 100
```

### Question 6
Drop rows containing missing values.

```python
df.dropna()
```

### Question 7
Drop columns containing missing values.

```python
df.dropna(axis=1)
```

### Question 8
Replace missing numerical values using mean.

```python
df_mean = df.copy()
num_cols = df_mean.select_dtypes(include=np.number).columns
df_mean[num_cols] = df_mean[num_cols].fillna(df_mean[num_cols].mean())
df_mean
```

### Question 9
Replace missing numerical values using median.

```python
df_median = df.copy()
df_median[num_cols] = df_median[num_cols].fillna(df_median[num_cols].median())
df_median
```

### Question 10
Replace missing categorical values using mode.

```python
df_mode = df.copy()

cat_cols = df_mode.select_dtypes(include=['object', 'string']).columns

for col in cat_cols:
    df_mode[col] = df_mode[col].fillna(df_mode[col].mode()[0])

df_mode
```
