### Question 1
Create one dataframe for the student record of 10 students with student basic information such as Rollno, name, age, Department.

```python
import pandas as pd

# ---------- FAKE DATA FOR 10 STUDENTS ----------
data = {
    "Rollno": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Rahul", "Ananya", "Vikram", "Sneha", "Arjun",
             "Priya", "Karan", "Neha", "Rohan", "Ishita"],
    "Age": [20, 21, 19, 22, 20, 23, 21, 20, 22, 19],
    "Department": ["CSE", "ECE", "ME", "CSE", "EE",
                   "IT", "CSE", "ECE", "ME", "IT"]
}

df = pd.DataFrame(data)

print("\nDATAFRAME:\n")
print(df)
```

### Question 2
Display starting 5 rows and end 5 rows.

```python
# ---------- HEAD & TAIL ----------
k = int(input("\nHow many rows to show from start & end? "))
print("\nFirst rows:\n", df.head(k))
print("\nLast rows:\n", df.tail(k))
```

### Question 8
Print two columns.

```python
# ---------- TWO COLUMNS ----------
print("\nAvailable Columns:")
print(list(df.columns))
c1 = input("Enter first column: ")
c2 = input("Enter second column: ")
print(df[[c1, c2]])
```

### Question 9
Print one and multiple rows.

```python
# ---------- ROW SELECTION ----------
opt = input("\nSingle row or multiple rows? (single/multiple): ")

if opt == "single":
    idx = int(input("Enter row index: "))
    print(df.loc[idx])
else:
    s = int(input("Start index: "))
    e = int(input("End index: "))
    print(df.loc[s:e])
```

### Question 10
Print row based on condition.

```python
# ---------- CONDITIONAL ROW ----------
print("\nAvailable Columns:")
print(list(df.columns))
cond_col = input("Enter column for condition: ")
val = int(input("Enter value (greater than): "))
print(df[df[cond_col] > val])
```

### Question 11
Add a new column.

```python
# ---------- ADD COLUMN ----------
new_col = input("\nEnter new column name: ")
new_val = input("Enter value for all rows: ")
df[new_col] = new_val
print("\nAfter adding column:\n", df)
```

### Question 12
Delete a column.

```python
# ---------- DELETE COLUMN ----------
print("\nAvailable Columns:")
print(list(df.columns))
del_col = input("Enter column to delete: ")
df.drop(del_col, axis=1, inplace=True)
print("\nAfter deleting column:\n", df)
```
