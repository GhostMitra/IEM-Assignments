### Question 1
Create a patient dataset using Pandas with features such as Age, Gender and treatment cost.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create patient dataset
data = {
    "Patient_ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Age": [25, 40, 32, 55, 60, 45, 28, 50],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Treatment_Cost": [12000, 18000, 15000, 22000, 25000, 20000, 13000, 23000]
}

df = pd.DataFrame(data)
df
```

### Question 2
Plot a bar chart showing patient-wise treatment cost.

```python
plt.figure()
plt.bar(df["Patient_ID"], df["Treatment_Cost"])
plt.xlabel("Patient ID")
plt.ylabel("Treatment Cost")
plt.title("Patient-wise Treatment Cost")
plt.grid(True)
plt.show()
```

### Question 3
Plot a line graph showing patient age distribution.

```python
plt.figure()
plt.plot(df["Patient_ID"], df["Age"])
plt.xlabel("Patient ID")
plt.ylabel("Age")
plt.title("Patient Age Distribution")
plt.grid(True)
plt.show()
```

### Question 4
Plot a histogram of patient ages.

```python
plt.figure()
plt.hist(df["Age"])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution Histogram")
plt.grid(True)
plt.show()
```

### Question 5
Plot a pie chart showing gender-wise patient distribution.

```python
gender_count = df["Gender"].value_counts()
plt.figure()
plt.pie(gender_count, labels=gender_count.index, autopct="%1.1f%%")
plt.title("Gender-wise Patient Distribution")
plt.show()
```

### Question 6
Plot a scatter plot between age and treatment cost.
### Question 7
Add title, labels, and grid to a plot.

```python
plt.figure()
plt.scatter(df["Age"], df["Treatment_Cost"])
plt.xlabel("Age")
plt.ylabel("Treatment Cost")
plt.title("Age vs Treatment Cost")
plt.grid(True)
plt.show()
```

### Question 8
Plot a box plot for treatment cost.

```python
plt.figure()
plt.boxplot(df["Treatment_Cost"])
plt.ylabel("Treatment Cost")
plt.title("Treatment Cost Distribution")
plt.grid(True)
plt.show()
```
