### Question 1
Create a line plot showing the variation of Math marks for all students with respect to Student_ID.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample student dataset
data = {
    "Student_ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Math": [78, 85, 67, 90, 72, 88, 60, 95],
    "Science": [82, 79, 70, 92, 75, 85, 65, 93],
    "English": [75, 80, 68, 88, 70, 83, 62, 90],
    "Attendance": [85, 90, 78, 95, 80, 92, 75, 98],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"]
}

df = pd.DataFrame(data)

print(df)

# 1 Line plot (Math vs Student_ID)
plt.figure()
plt.plot(df["Student_ID"], df["Math"])
plt.xlabel("Student ID")
plt.ylabel("Math Marks")
plt.title("Math Marks Variation")
plt.show()
```

### Question 2
Plot a bar chart to compare Science marks of all students.

```python
plt.figure()
plt.bar(df["Student_ID"], df["Science"])
plt.xlabel("Student ID")
plt.ylabel("Science Marks")
plt.title("Science Marks Comparison")
plt.show()
```

### Question 3
Draw a histogram for English marks and analyze the distribution.

```python
plt.figure()
plt.hist(df["English"])
plt.xlabel("English Marks")
plt.ylabel("Frequency")
plt.title("English Marks Distribution")
plt.show()
```

### Question 4
Create a scatter plot between Attendance (%) and Math marks.

```python
plt.figure()
plt.scatter(df["Attendance"], df["Math"])
plt.xlabel("Attendance (%)")
plt.ylabel("Math Marks")
plt.title("Attendance vs Math Marks")
plt.show()
```

### Question 5
Plot a box plot for Math, Science, and English marks to identify score variations and outliers.

```python
plt.figure()
plt.boxplot([df["Math"], df["Science"], df["English"]])
plt.xticks([1, 2, 3], ["Math", "Science", "English"])
plt.title("Marks Distribution")
plt.show()
```

### Question 6
Draw a pie chart representing the gender distribution of students.

```python
gender_counts = df["Gender"].value_counts()
plt.figure()
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

### Question 7
Create a stacked bar chart showing Math, Science, and English marks for each student.

```python
plt.figure()
plt.bar(df["Student_ID"], df["Math"])
plt.bar(df["Student_ID"], df["Science"], bottom=df["Math"])
plt.bar(df["Student_ID"], df["English"], bottom=df["Math"] + df["Science"])
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.title("Stacked Marks Chart")
plt.show()
```

### Question 8
Plot a horizontal bar chart showing average marks per subject.

```python
avg_marks = df[["Math", "Science", "English"]].mean()
plt.figure()
plt.barh(avg_marks.index, avg_marks.values)
plt.xlabel("Average Marks")
plt.title("Average Marks Per Subject")
plt.show()
```
