### Question 1
Write a Python program to create a Student dataset with the following columns:

(math_marks, science_marks, english_marks). 

Insert at least 10 student records with different value ranges.

```python
import pandas as pd

data = {
    "student_id": range(1,11),
    "math_marks": [45,78,90,66,55,88,92,73,60,81],
    "science_marks": [50,85,88,70,58,90,95,75,65,83],
    "english_marks": [48,80,86,68,57,89,94,74,63,82]
}

df = pd.DataFrame(data)
df
```

### Question 2
Write a Python program to apply Min–Max normalization to the dataset without using sklearn

```python
minmax_manual = df.copy()

for col in ["math_marks","science_marks","english_marks"]:
    min_val = df[col].min()
    max_val = df[col].max()
    minmax_manual[col] = (df[col] - min_val) / (max_val - min_val)

minmax_manual
```

### Question 3
Normalize the dataset using MinMaxScaler from sklearn.preprocessing and display the transformed data.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df.iloc[:,1:])

minmax_sklearn = pd.DataFrame(
    scaled,
    columns=["math_marks","science_marks","english_marks"]
)

minmax_sklearn
```

### Question 4
Write a Python program to standardize the student dataset manually using the Z-score.

```python
z_manual = df.copy()

for col in ["math_marks","science_marks","english_marks"]:
    z_manual[col] = (df[col] - df[col].mean()) / df[col].std()

z_manual
```

### Question 5
Apply StandardScaler on the dataset using scikit-learn and compare the output with the manual standardization.

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
std_scaled = std_scaler.fit_transform(df.iloc[:,1:])

std_sklearn = pd.DataFrame(
    std_scaled,
    columns=["math_marks","science_marks","english_marks"]
)

std_sklearn
```

### Question 6
Write a Python program to plot:
- Original values
- Min–Max normalized values
- Standardized values

for the Maths subject using matplotlib.

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df["math_marks"])
plt.xlabel("Student Index")
plt.ylabel("Math Marks")
plt.title("Original Math Marks")
plt.show()

plt.figure()
plt.plot(minmax_manual["math_marks"])
plt.xlabel("Student Index")
plt.ylabel("Normalized Value")
plt.title("Min-Max Normalized Math Marks")
plt.show()

plt.figure()
plt.plot(std_sklearn["math_marks"])
plt.xlabel("Student Index")
plt.ylabel("Standardized Value")
plt.title("Standardized Math Marks")
plt.show()
```
