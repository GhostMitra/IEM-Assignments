### Question 1
Create a student data frame using missing values. The dataset should have the following information.

('ID','Name','Age',’Gender’,'Math_Marks','Science_Marks',’English_Marks’,’Percentage of Attendance’)

```python
import pandas as pd
import numpy as np

data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Amit', 'Riya', 'Sohan', 'Neha', 'Arjun'],
    'Age': [20, np.nan, 22, np.nan, 21],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Math_Marks': [85, np.nan, 78, 90, np.nan],
    'Science_Marks': [88, 92, np.nan, 85, 80],
    'English_Marks': [75, 80, 82, np.nan, np.nan],
    'Percentage of Attendance': [90, 85, np.nan, 88, 92]
}

df = pd.DataFrame(data)
df
```

### Question 2
Check whether missing values are there or not.

```python
df.isnull()
```

### Question 3
Print the column which is having missing values

```python
df.columns[df.isnull().any()]
```

### Question 4
Filling missing ‘Age’ value by mean value.

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

### Question 5
Fill Missing Marks with Subject-wise Mean.

```python
marks = ['Math_Marks', 'Science_Marks', 'English_Marks']
df[marks] = df[marks].fillna(df[marks].mean())
df
```

### Question 6
Drop Rows with Missing Value

```python
df.dropna()
```
